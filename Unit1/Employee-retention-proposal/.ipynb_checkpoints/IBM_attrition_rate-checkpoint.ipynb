## Fixing data

There are some aspects of an employees life that IBM cannot control (such as how far a commute to work is), therefore, I will drop all columns that would not be useful in helping IBM retain their employees. Though there are some parameters that may be useful to consider in future hiring, I will drop those as well if it's not helpful information in keeping those already employed. I will leave some variables that are known to be a source of discrimination in the workplace, such as age.

```python
df = df.loc[:, ['Age', 'Attrition', 'BusinessTravel', 'DailyRate', 'Department',
       'EnvironmentSatisfaction', 'Gender', 'HourlyRate',
       'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction', 'RelationshipSatisfaction',
       'MonthlyIncome', 'MonthlyRate',
       'OverTime', 'PercentSalaryHike', 'PerformanceRating',
       'StockOptionLevel', 
       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']]
```

```python
df.describe()
```

## Explore retention rates

```python
left = df[df.Attrition == 'Yes']
retained = df[df.Attrition=='No']
```

```python
left.describe()
```

```python
retained.describe()
```

```python
print("Attrition rates of IBM employees: ", len(left)/len(retained))
```

```python
depts = df.Department.unique()
print("Departments included: ", depts)
```

```python
print("Attrition rates per department: ")
for dept in depts:
    print(dept, ":  ", len(left[left.Department == dept]) / len(df[df.Department == dept]))
```

Interesting, worthwhile to see why R&D may have significantly better retention than Sales and perhaps try to model that environment

```python
print("Attrition rates by gender: ")
print("Male: ", len(left[left.Gender == 'Male']) / len(df[df.Gender == 'Male']))
print("Female: ", len(left[left.Gender=='Female']) / len(df[df.Gender=='Female']))
```

Do a ttest//mannwhitney to see if this is significant

```python
print("Retention rates by happiness with colleagues (between 1-4): ")
print("Attrition: ", left.RelationshipSatisfaction.mean())
print("Retention: ", retained.RelationshipSatisfaction.mean())
```

Is this one significant?

```python
sns.heatmap(df.corr(),
           xticklabels=df.corr().columns,
           yticklabels=df.corr().columns)

plt.show()
```

```python
sns.boxplot(x='Attrition', y='DailyRate', data=df)
```

```python
sns.boxplot(x='Attrition', y='YearsSinceLastPromotion', data=df)
plt.show()
```

This was unexpected - there is a greater spread of time since last promotion in the retained group, indicating recent promotions is not a good attrition prevention method.

```python
sns.boxplot(x='Attrition', y='YearsWithCurrManager', data=df)
plt.show()
```

Bingo - perhaps focusing on manager retention will prevent general employee attrition rates.
