## Challenge: what do you see?
In this assignment, I will select the training data from Kaggle's Titanic dataset to generate a few visualizations describing the passengers.

Data source: https://www.kaggle.com/c/titanic/data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (10.0, 5.0)
plt.rcParams.update({'font.size': 22})
```

```python
df = pd.read_csv('titanic-train.csv')
df.head()
```

```python
df.columns
```

```python
# dropping any passenger data with NaN in relevant columns
df = (df.dropna(subset=['Survived', 'Pclass', 'Age', 'Fare', 'Embarked']))
```

```python
df.describe()
```

```python
plt.subplot(2, 1, 1)
df.Pclass.value_counts().plot(kind='bar', figsize=(20,10))
plt.xlabel("Ticket class")
plt.ylabel("Count")
plt.title("Distribution of Passenger Ticket Class")

plt.subplot(2, 1, 2)
plt.scatter(x=df['Age'], y=df['Fare'])
plt.xlabel("Age")
plt.ylabel("Ticket cost ($)")
plt.title("Ticket cost by age")
plt.tight_layout()
plt.show()
```

Analysis: Most of the tickets are third class. Surprisingly, there are more first class than second class tickets. The highest prices paid seem to be by passengers somewhere between the ages of 18 and 45.

```python
plt.boxplot(df['Fare'])
plt.ylabel("Ticket cost ($)")
plt.title("Boxplot of Ticket Prices")
plt.show()
```

Analysis: There is a wide range of fares paid by passengers. The highest paid fare is by a single person at roughly 500USD. 
Interestingly, The next highest costs are just over half of that, at around 275USD. That is quite the outlier, particularly when the majority of the fares fall under 100USD.

```python
df.Survived.value_counts().plot(kind='bar')
plt.xlabel("Survival (0 = no, 1 = yes)")
plt.ylabel("Count")
plt.title("Passenger survival")
plt.show()
```

Analysis: Unfortunately, more passengers died than survived. Over 500 passengers died on the Titanic that day; less than 300 survived.

```python
sns.lmplot(x='Age', y='Fare', hue='Survived', data=df, fit_reg=False, scatter_kws={'alpha':0.5})
plt.show()
```

Analysis: younger passengers were more likely to live as there is a greater chance for those roughly below the age of 40. However, those who paid more for their fare are more likely to live, regardless of age. Interestingly, the oldest person on the Titanic survived, despite purchasing a low-cost fare. This was an exception to the rule that preferences were given to the young and/or wealthy.
