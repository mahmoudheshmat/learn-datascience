# Purpose:
To determine correlation in lifestyle habits leading to cervical cancer

## Source:
The dataset was collected at 'Hospital Universitario de Caracas' 
in Caracas, Venezuela. The dataset comprises demographic information, 
habits, and historic medical records of 858 patients. Several patients 
decided not to answer some of the questions because of privacy concerns 
(missing values).
http://archive.ics.uci.edu/ml/datasets/Cervical+cancer+%28Risk+Factors%29

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('risk_factors_cervical_cancer.csv', na_values=['?'])
print(df)
```

```python
# show a histogram of age groups of participants
plt.hist(df['Age'], color='purple', bins=40)
plt.xlabel('Age')
plt.ylabel('Distribution')
plt.title("Age Distribution of Participants")

'''add a best fit curve with mean and std calculated below
bins = 40
n, bins, patches = (df['Age'], bins, density=1)
fig, ax = plt.subplots()

mean = 26.9
std = 8.5
bestfit = ((1 / (np.sqrt(2*np.pi) * std)) * np.exp(-0.5 * (1 / std * (bins - mean))**2))
ax.plot(bins, bestfit, 'x')

fig.tight_layout()'''
plt.show()
```

```python
promiscuity = df['Age']
plt.boxplot(promiscuity)
plt.ylabel('Age of survey participants')
plt.show()
```

## Conclusion: 
The average age of the sample population is 26.8, with a standard deviation of 8.5 years. The youngest woman surveyed was 13 years old, the oldest at 84 years of age.
