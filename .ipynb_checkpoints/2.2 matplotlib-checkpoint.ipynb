```python
np.random.seed(1221) # random is consistent and replicable
```

```python
df = pd.DataFrame() # blank data frame
```

```python
# column of random numbers b/w 0-1
df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2
```

```python
# transformations on index to create extra columns
df['counts_sq'] = df.index ** 2
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
# by default, index is x-axis
plt.plot(df['rand'])
plt.show()
```

```python
plt.plot(df['rand_sq'])
plt.show()
```

```python
# customize plot
plt.plot(df['rand'], color='red')
plt.ylim([-0.01, 1.5])
plt.ylabel('Values')
plt.xlabel('Indices')
plt.title('Random Series')
plt.show()
```

```python
plt.plot(df['rand'], color='purple')
plt.plot(df['rand_shift'], color='green')
plt.ylim([-0.1, 3.1])
plt.ylabel("Values")
plt.title("Random Series")
plt.show()
```

```python
plt.scatter(x=df['rand'], y=df['rand_sq'])
plt.show()
```

```python
plt.scatter(
    x=df['rand'],
    y=df['rand_sq'],
    color='purple',
    marker='x', s=10
)
plt.scatter(
    x=df['rand'],
    y=df['rand_shift'],
    color='green',
    marker='o', s=30
)
plt.show()
    
```

## Subplots

```python
# assuming matplotlib(pyplot), numpy, and pandas are imported
np.random.seed(1227)

df = pd.DataFrame()
df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2

df['counts_sq'] = df.index ** 2
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
# generate multiple plots in a single figure
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['rand'], color='purple')
plt.ylabel('Values')
plt.title('Random Series')

plt.subplot(1, 2, 2)
plt.plot(df['rand_shift'], color='green')
plt.ylabel('Shifted Values')
plt.title('Shifted Series')
plt.show()
```

```python
# adding 'tight_layout

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['rand'], color='purple')
plt.ylabel('Values')
plt.title('Random Series')

plt.subplot(1, 2, 2)
plt.scatter(x = df['rand_sq'], y = df['rand'], color='green')
plt.ylabel('Squared Values')
plt.title('Squared Series')

plt.tight_layout()
plt.show()
```
