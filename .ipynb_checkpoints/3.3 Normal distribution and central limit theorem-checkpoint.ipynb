## Normal distribution
If:
- mean = median
- data tapers symmetrically to the left and right
- info can be summarized with mean (mu) and std dev (sigma)
- ~68% of values fall within 1 std dev, 95% within 2, 99.7% within 3
- area under a curve is 1
- percentiles and z-scores, as well as many stats tests like t-test and ANOVA assume normal distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# generate normal distribution with python
mean = 0
sd = 1
n = 1000

df = pd.DataFrame({'rand': np.random.normal(mean, sd, n)})

df.hist()
plt.show()
```

## Is it normal?

Best method of deciding whether data is normally distributed is to inspect data with
- histograms
- quantile-quantile (QQ) plots

### QQ plots
- graph a variable with an unknown distribution against a variable with known distribution
    - values sorted in ascending order, plotted against each other
    - x axis = known variable
    - y axis = unknown variable
- if the distributions are the same, the result should be a straight line from origin to top right corner of plot

```python
# example: testing if data is normally distributed
rand1 = np.random.normal(50, 300, 1000)
rand2 = np.random.poisson(1, 1000)

# sorting values in ascending order
rand1.sort()
rand2.sort()

# our comparison, normal, distribution:
norm = np.random.normal(0, 1, 1000)

norm.sort()
```

```python
# plotting rand1 vs norm in qqplots
plt.plot(norm, rand1, "o")
plt.show()
```

```python
# plotting rand2 vs norm in qq plots
plt.plot(norm, rand2, "o")
plt.show()
```

### Analysis:
Looking at QQ plot, rand1 is normally distributed, while rand2 is not (it is Poisson)

```python
plt.hist(rand1, bins=20, color='c')

# add a vertical line at the mean
plt.axvline(rand1.mean(), color='blue', linestyle='solid', linewidth=2)

# add a vertical line at one std dev above the mean
plt.axvline(rand1.mean() + rand1.std(), color='b', linestyle='dashed', linewidth=2)

# add a vertical line at one std dev below the mean
plt.axvline(rand1.mean() - rand1.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
# plot the same histogram for rand2
plt.hist(rand2, bins=20, color='c')

# add vertical line at the mean
plt.axvline(rand2.mean(), color='blue', linestyle='solid', linewidth=2)

# add upper 1 std dev line
plt.axvline(rand2.mean() + rand2.std(), color='blue', linestyle='solid', linewidth=2)

# add lower 1 std dev line
plt.axvline(rand2.mean() - rand2.std(), color='b', linestyle='solid', linewidth=2)

plt.show()
```

From the graphs above, it's now evident that rand1 is a normal distribution, with 50% of the data points on either side of the mean, and 67% of the data within one standard deviation of that mean. Conversely, rand2 is not a normal distribution, and has its mean where data clusters, but it is not a measure of 'central tendency' anymore.

## Bernoulli distribution
- Represents 2 possible outcomes of an event (i.e. coin flip)
- when a distribution is
    - discrete: it has probability mass function
    - continuous: it has a probability density function

where p is probability of outcome (k)

```python
# generate a Bernoulli distribution, p=0.5
bernoulli= np.random.binomial(1, 0.5, 1000)

# plot a histogram
plt.hist(bernoulli)

# print the histogram
plt.show()
```

```python
# generate a binomial distribution with n= 20 and p=0.5
binomial = np.random.binomial(20, 0.5, 1000)

# plot a histogram
plt.hist(binomial)

# print the histogram
plt.show()
```

## Gamma distribution
- Represents the time until an event where the event begins unlikely, becomes more likely, then becomes less likely again
    - ex. lifespan until death: few die in their youth, starts to become likely in old age, then less likely because there are few people left to die in extreme old age b/c they've already all died

```python
# generate gamma distribution with shape=5 and scale=1
gamma = np.random.gamma(5, 1, 1000)

# plot a histogram
plt.hist(gamma)

plt.show()
```

## Poisson distribution
- number of times a given eventw ill occur during a given time interval
    - ex. radio show phone calls
- lambda: the rate that events occur during a given time period

```python
# generate a Poisson distribution with lambda = 3
poisson = np.random.poisson(3, 1000)

# plot a histogram
plt.hist(poisson)
plt.show()
```

## Conditional distribution
- "filtering" for a specific subset of data, with the qualifier of the subset being the condition
    - ex. looking at a distribution of the amount of time consumers have spent on an ecommerce site, including the condition that the customers 
    

```python
# create a data frame for simulated ecommerce data
# normally distributed with mean = 75, stdev = 25

ecommerce = pd.DataFrame()
ecommerce['spending'] = np.random.normal(75, 25, 1000)

# plot a histogram
plt.hist(ecommerce['spending'])
plt.show()
```

```python
# adding a variable to use as a conditional
# number of times a user visits the site
ecommerce['visit_count'] = np.random.randint(0, 5, 1000)

# selecting only the cases where the visit count is greater than 2
plt.hist(ecommerce[ecommerce['visit_count'] > 2]['spending'])
plt.show()
```
