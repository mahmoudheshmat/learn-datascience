```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# generate 1000 random values
# with a "center" of 10 and a standard deviation of 5
x = np.random.normal(10, 5, 1000)

# plot as histogram
plt.hist(x)
plt.show()

# by default, matplotlib distributes values into 10 bins
```

```python
# to adjust bin upper/lower limits and number
plt.hist(x, bins=40, color='red')
plt.title('Default Bin Placement Demo')
plt.xlabel('Random Values')
plt.ylabel('No. of occurences')
plt.show()
```

```python
# pass an array of specific bounds for bin location
x = np.random.normal(10, 5, 1000)
y = np.random.normal(15, 5, 10000)

# setting bins at integer markers between -10 and 40
# alpha sets opacity of color
plt.hist(x, color='blue', bins=np.arange(-10, 40), alpha=0.5)
plt.hist(y, color='red', bins=np.arange(-10, 40), alpha=0.5)
plt.title('Manually setting bin placement')
plt.xlabel('Random Values')

plt.show()
```

```python
# same data above, normalized:
plt.hist(x, density=True, color='blue', bins=np.arange(-10, 40), alpha=0.5)
plt.hist(y, density=True, color='red', bins=np.arange(-10, 40), alpha=0.5)
plt.title('Normalized histograms')
plt.xlabel('Random Values')

plt.show()
```
