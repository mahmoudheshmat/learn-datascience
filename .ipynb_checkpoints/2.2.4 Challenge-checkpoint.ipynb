### In this notebook, I will generate a set of visuals around a selected data using pyplot
Data obtained from the 500 Cities project on data.gov

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.rcParams.update({'font.size': 22})
```

```python
# import dataframe
df = pd.read_csv("https://chronicdata.cdc.gov/resource/5hnu-2c66.csv")
```

```python
df.columns
```

```python
df = df.drop(columns=[':@computed_region_bxsw_vy29', ':@computed_region_he4y_prf8', 'datavaluefootnote',
                 'datavaluefootnotesymbol', 'stratification2', 'stratification3', 'stratificationcategory2',
                 'stratificationcategory3', 'stratificationid2', 'stratificationid3', 'geolocation', 'geolocation_address',
                 'geolocation_city', 'geolocation_state', 'geolocation_zip', 'highconfidencelimit', 'lowconfidencelimit', 
                 'stratificationcategory2', 'stratificationcategory3', 'stratification2', 'stratification3', 'response',
                 'responseid', 'datasource', 'stratificationcategoryid1', 'stratificationid1', 'stratificationid2', 
                 'stratificationid3'])
```

```python
df.columns
```

```python
df.head()
```

```python
df = df.dropna(subset= (['datavaluealt']))
```

```python
df.describe()
```

```python
# questions asked relative to disease
df.topic.value_counts().plot(kind = 'bar')
plt.xlabel('Topic')
plt.ylabel('Counts')
plt.title('Questions asked per chronic disease')
plt.show()
```

Data is not just from the 50 states, but includes Washington DC and territories:

```python
pd.Series.unique(df['locationdesc'])
```

```python
df['locationdesc'].value_counts()
```

```python
location_counts = df['locationabbr'].value_counts()
```

```python
location_counts.plot(kind='bar')
plt.xlabel('Location')
plt.ylabel('Counts')
plt.title('Location of respondents')
plt.show()
```

```python
strat_cat_counts = df['stratificationcategory1'].value_counts()
```

```python
strat_cat_counts.plot(kind='bar')
plt.xlabel('Stratification categories')
plt.ylabel('Counts')
plt.title('Differentiation categories counts')
plt.show()
```

```python
strat_counts = df['stratification1'].value_counts()
strat_counts.plot(kind='bar')
plt.xlabel('Stratification types')
plt.ylabel('Counts')
plt.title('Counts across stratification types')
plt.show()
```
